# Decision tree model

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()

dt.fit(X_train,Y_train)

dt_ac=accuracy_score(Y_test,Y_pred)
Y_pred=dt.predict(X_test)

recall = recall_score(Y_test,Y_pred, average='macro')
precision = precision_score(Y_test,Y_pred, average='macro')
score = f1_score(Y_test,Y_pred, average='macro')
print('Accuracy: %.3f' % dt_ac)
print('Precision: %.3f' % precision)
print('Recall: %.3f' % recall)
print('F-Measure: %.3f' % score)
```

# 2) KNN

```python
#### We need to use Standard Scaler for knn since distance matter here
```

```python
X_train_std=std.fit_transform(X_train)

X_test_std=std.transform(X_test)
```

# KNN

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train_std,Y_train)
Y_pred=knn.predict(X_test_std)
knn_ac=accuracy_score(Y_test,Y_pred)
recall = recall_score(Y_test,Y_pred, average='macro')
precision = precision_score(Y_test,Y_pred, average='macro')
score = f1_score(Y_test,Y_pred, average='macro')


print('Accuracy: %.3f' % dt_ac)
print('Precision: %.3f' % precision)
print('Recall: %.3f' % recall)
print('F-Measure: %.3f' % score)

```

# 3) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
lr=LogisticRegression()
lr.fit(X_train_std,Y_train)
Y_pred=lr.predict(X_test_std)

lr_ac=accuracy_score(Y_test,Y_pred)
recall = recall_score(Y_test,Y_pred, average='macro')
precision = precision_score(Y_test,Y_pred, average='macro')
score = f1_score(Y_test,Y_pred, average='macro')


print('Accuracy: %.3f' % dt_ac)
print('Precision: %.3f' % precision)
print('Recall: %.3f' % recall)
print('F-Measure: %.3f' % score)
```

```python
import scipy.stats as stats
import numpy as np
import math
sample = np.random.choice(a=Y_pred,size = 11)
sample_mean = sample.mean()
z_critical = stats.norm.ppf(q = 0.95)
pop_stdev = Y_pred.std()
margin_of_error = z_critical * (pop_stdev/math.sqrt(250))
confidence_interval = (sample_mean - margin_of_error,sample_mean +
margin_of_error)
print("Confidence interval:",end=" ")
print(confidence_interval)
```
